$$|\psi_A\rangle \otimes |\psi_B\rangle = \begin{pmatrix} \alpha_0\cdot |\psi_B\rangle \\ \alpha_1\cdot |\psi_B\rangle\end{pmatrix}=\begin{pmatrix} \alpha_0\beta_0 \\ \alpha_0\beta_1 \\ \alpha_1\beta_0 \\ \alpha_1\beta 1\end{pmatrix}$$
