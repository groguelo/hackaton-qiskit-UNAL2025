Dise√±a un or√°culo de fase de 2 qubits que marque espec√≠ficamente el estado $|10\rangle$ (qubit 1 es $|1\rangle$, qubit 0 es $|0\rangle$) con una fase de -1.

### El Principio de Transformaci√≥n de Base

La compuerta CZ solo aplica la fase de -1 cuando ambos qubits son $|1\rangle$. Para apuntar a un estado diferente como $|10\rangle$, debemos:

1.  Usar compuertas Pauli $X$ para voltear el estado objetivo ($|10\rangle$) al estado $|11\rangle$.
2.  Aplicar la compuerta CZ.
3.  Usar compuertas $X$ subsecuentes para devolver el vector de estado a la base original.

### Requisitos:
1.  **Construcci√≥n del Circuito:** Crea un `QuantumCircuit` de 2 qubits.
2.  **Objetivo $|10\rangle$:** Aplica la secuencia necesaria de compuertas $X$, una compuerta $CZ$, y las compuertas $X$ inversas para marcar *solo* el estado $|10\rangle$.
3.  **Verificaci√≥n:** Usa el simulador `Statevector` para confirmar que evolucionar el estado de entrada $|10\rangle$ a trav√©s de tu circuito resulta en el estado final $-|10\rangle$.

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
import numpy as np

# --- 1. Construcci√≥n del Or√°culo para |10> ---
# Crea un circuito con 2 qubits
oracle_circuit = QuantumCircuit(2)

# El objetivo es crear un or√°culo que voltee la fase del estado objetivo |10>.

# YOUR CODE HERE
raise NotImplementedError()

print("Circuito del Or√°culo para |10>:")
print(oracle_circuit.draw('text'))

# --- 2. Verificaci√≥n ---

# Define el estado de entrada |10>. En el orden de Qiskit |q1q0>, este es el tercer elemento
# en el vector de estado: |00>, |01>, |10>, |11>
input_state_10 = Statevector([0, 0, 1, 0])
print(f"\nEstado Inicial |10>: {np.round(input_state_10.data, 2)}")

# Evoluciona el estado a trav√©s del circuito para ver el efecto del or√°culo
final_state = input_state_10.evolve(oracle_circuit)

# El estado final esperado es -|10> = [0, 0, -1, 0]
expected_phase_flip = np.array([0, 0, -1, 0])

print(f"Estado Final:        {np.round(final_state.data, 2)}")

# Verifica si el estado final coincide con el estado esperado con un volteo de fase
if np.allclose(final_state.data, expected_phase_flip):
    print("\n‚úÖ ¬°√âxito! La fase del estado |10> fue volteada a -|10>.")
else:
    print("\n‚ùå Verificaci√≥n fallida. El estado no fue volteado correctamente.")
```
